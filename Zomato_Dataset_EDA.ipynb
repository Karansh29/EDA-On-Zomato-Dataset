```python
df_country=pd.read_excel('Country-Code.xlsx')
```

```python
df_country.head()
```

```python
df.columns
```

```python
#merging two columns
final_df=pd.merge(df,df_country,on='Country Code', how='left')
```

```python
final_df.head()
```

```python
#check datatype
final_df.dtypes
```

```python
final_df.columns
```

```python
final_df.Country.value_counts()
```

```python
country_name=final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
```

```python
##Pie Chart--Top 3 Country
plt.pie(country_val[:3],labels=country_name[:3],autopct='%1.2f%%')
```

Observation: Zomato Maximum Records or Transaction are from India

After that USA then United Kingdoms

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})  ##rename column name
```

```python
ratings
```

Observation

1. When Rating is between 4.5  to 4.9 ---> Excellent
2. When Rating is between 4.0  to 4.4 ---> Very Good
3. When Rating is between 3.5  to 3.9 ---> Good
4. When Rating is between 2.5  to 3.4 ---> Average
5. When Rating is between 0.0  to 2.4 ---> Poor

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] =(12,6)
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings)
```

```python
#mapping the colors
sns.barplot(x="Aggregate rating",y="Rating Count", hue='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
```

Observation

1. Not Rated count is very high
2. Maximum number  of rating are between 2.5 to 3.4

```python
## Count Plot
sns.countplot(x= "Rating color",data = ratings,palette=['blue','red','orange','yellow','green','green'])
```
